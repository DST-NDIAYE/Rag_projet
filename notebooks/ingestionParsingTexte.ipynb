# lecture fichier txt 

```python
with open("../data/texte/intro_rag.txt" , "r" , encoding="utf-8") as f :
    texte  = f.read()
```

```python
print(texte)
```

#  texte to doc

```python
from langchain_core.documents import Document
```

```python
doc = Document(
    page_content=texte , 
    metadata = { "source" : "../data/texte/intro_rag.txt" ,
                "hoteur" : "Serigne", 
                "format" : "txt"
                }
)
```

```python
print(doc.page_content)
```

```python
print(doc.metadata["source"])
```

# nettoyage de texte

```python
from langchain_community.document_loaders import TextLoader
```

```python
loader = TextLoader( "../data/texte/texte_exemple_rag.txt" , encoding="utf-8" )
```

```python
doc2 = loader.load() 
```

```python
print(doc2[0].page_content[:300])
```

```python
texte2 = doc2[0].page_content
```

```python
import re

def clean_texte(texte) :
    
    txt = texte.lower()
    txt = re.sub(r"[^\w\s]",  "" , txt )
    txt = txt.strip()

    return txt
```

```python
print(clean_texte(texte2))
```
