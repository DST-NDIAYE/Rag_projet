# lecture fichier txt 

```python
with open("../data/texte/intro_rag.txt" , "r" , encoding="utf-8") as f :
    texte  = f.read()
```

```python
print(texte)
```

#  texte to doc

```python
from langchain_core.documents import Document
```

```python
doc = Document(
    page_content=texte , 
    metadata = { "source" : "../data/texte/intro_rag.txt" ,
                "hoteur" : "Serigne", 
                "format" : "txt"
                }
)
```

```python
print(doc.page_content)
```

```python
print(doc.metadata["source"])
```
